```python
"""
InstructPix2Pix Image Editing Pipeline
=======================================
Applies text-guided image edits using the InstructPix2Pix model from Diffusers.
Example: "make the car look like it's in the snow"

Requirements:
    pip install diffusers transformers accelerate torch Pillow

GPU recommended (runs on CPU but slowly).
"""

import sys
from pathlib import Path

import torch
from diffusers import StableDiffusionInstructPix2PixPipeline, EulerAncestralDiscreteScheduler
from PIL import Image
from IPython.display import display


# ──────────────────────────────────────────────────────────────────────────────
# Model helpers
# ──────────────────────────────────────────────────────────────────────────────

MODEL_ID = "timbrooks/instruct-pix2pix"


def load_pipeline(device: str = "auto") -> StableDiffusionInstructPix2PixPipeline:
    """
    Load and return the InstructPix2Pix pipeline.

    Args:
        device: "cuda", "mps", "cpu", or "auto" (picks best available).

    Returns:
        Loaded pipeline ready for inference.
    """
    if device == "auto":
        if torch.cuda.is_available():
            device = "cuda"
        elif torch.backends.mps.is_available():
            device = "mps"
        else:
            device = "cpu"

    print(f"[pipeline] Loading model on device: {device}")

    dtype = torch.float16 if device in ("cuda", "mps") else torch.float32

    pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
        MODEL_ID,
        torch_dtype=dtype,
        safety_checker=None,   # disable NSFW filter for non-restricted use-cases
    )
    pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
    pipe = pipe.to(device)

    # Memory optimisation for low-VRAM GPUs
    if device == "cuda":
        pipe.enable_attention_slicing()

    return pipe


# ──────────────────────────────────────────────────────────────────────────────
# Core editing function
# ──────────────────────────────────────────────────────────────────────────────

def edit_image(
    pipe: StableDiffusionInstructPix2PixPipeline,
    input_image: Image.Image,
    instruction: str,
    *,
    num_inference_steps: int = 50,
    image_guidance_scale: float = 1.5,
    text_guidance_scale: float = 7.5,
    num_images: int = 1,
    seed: int | None = None,
) -> list[Image.Image]:
    """
    Apply a text instruction to edit an image.

    Args:
        pipe:                  Loaded InstructPix2Pix pipeline.
        input_image:           PIL Image to edit.
        instruction:           Natural-language edit, e.g. "make it snow".
        num_inference_steps:   Denoising steps (more = higher quality, slower).
        image_guidance_scale:  How strongly the output stays close to the input
                               (1.0 = ignore input, ↑ = more faithful to input).
        text_guidance_scale:   How strongly the output follows the text instruction.
        num_images:            Number of variations to generate.
        seed:                  Optional RNG seed for reproducibility.

    Returns:
        List of edited PIL Images.
    """
    generator = None
    if seed is not None:
        generator = torch.manual_seed(seed)

    # InstructPix2Pix works best at 512x512; resize while preserving aspect ratio
    input_image = _resize_to_fit(input_image, max_side=512)

    print(f"[pipeline] Editing image ({input_image.size}) with instruction: '{instruction}'")

    result = pipe(
        prompt=instruction,
        image=input_image,
        num_inference_steps=num_inference_steps,
        image_guidance_scale=image_guidance_scale,
        guidance_scale=text_guidance_scale,
        num_images_per_prompt=num_images,
        generator=generator,
    )

    return result.images


def _resize_to_fit(image: Image.Image, max_side: int = 512) -> Image.Image:
    """Resize image so its longest side equals max_side, keeping aspect ratio."""
    w, h = image.size
    scale = max_side / max(w, h)
    if scale >= 1.0:
        return image.convert("RGB")
    new_w = int(w * scale)
    new_h = int(h * scale)
    # Round to multiples of 8 (required by diffusion models)
    new_w = (new_w // 8) * 8
    new_h = (new_h // 8) * 8
    return image.resize((new_w, new_h), Image.LANCZOS).convert("RGB")


# ──────────────────────────────────────────────────────────────────────────────
# Convenience wrapper: file -> file
# ──────────────────────────────────────────────────────────────────────────────

def run(
    input_path: str,
    instruction: str,
    output_path: str | None = None,
    *,
    device: str = "auto",
    steps: int = 50,
    image_guidance: float = 1.5,
    text_guidance: float = 7.5,
    num_images: int = 1,
    seed: int | None = None,
) -> list[str]:
    """
    End-to-end helper: load image -> edit -> save.

    Args:
        input_path:      Path to the source image file.
        instruction:     Edit instruction.
        output_path:     Output file path (auto-generated if None).
        device:          Device override.
        steps:           Inference steps.
        image_guidance:  Image guidance scale.
        text_guidance:   Text guidance scale.
        num_images:      Number of output images.
        seed:            RNG seed.

    Returns:
        List of saved output file paths.
    """
    input_path = Path(input_path)
    if not input_path.exists():
        raise FileNotFoundError(f"Input image not found: {input_path}")

    image = Image.open(input_path)

    pipe = load_pipeline(device=device)

    edited_images = edit_image(
        pipe,
        image,
        instruction,
        num_inference_steps=steps,
        image_guidance_scale=image_guidance,
        text_guidance_scale=text_guidance,
        num_images=num_images,
        seed=seed,
    )

    saved_paths = []
    stem = input_path.stem
    suffix = input_path.suffix or ".png"

    for i, img in enumerate(edited_images):
        if output_path and num_images == 1:
            out = Path(output_path)
        else:
            tag = f"_edited_{i}" if num_images > 1 else "_edited"
            out = input_path.parent / f"{stem}{tag}{suffix}"

        img.save(out)
        print(f"[pipeline] Saved: {out}")
        saved_paths.append(str(out))

    return saved_paths


# ──────────────────────────────────────────────────────────────────────────────
# Colab-safe runner (no argparse / main)
# ──────────────────────────────────────────────────────────────────────────────

# Upload your image first using the cell:
#   from google.colab import files
#   uploaded = files.upload()
#   input_filename = list(uploaded.keys())[0]

# Load model
pipe = load_pipeline(device="cuda")

# Open uploaded image
image = Image.open(input_filename)

# Edit — change the instruction to whatever you like!
outputs = edit_image(
    pipe,
    image,
    "make the car red in color",
    seed=42
)

# Show inline & save
display(outputs[0])
outputs[0].save("result.jpg")
print("✅ Saved as result.jpg")
```
